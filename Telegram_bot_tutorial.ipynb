#Criando um método para salvar a imagem temporariamente no google drive

```
def salvarImagem(url,nomeArquivo):
  import urllib.request
  import os
  import time
  urllib.request.urlretrieve(url,nomeArquivo)
  time.sleep(2)
  if os.path.isfile(nomeArquivo):
    print('arquivo [{}] salvo com sucesso'.format(nomeArquivo))
  
  

```

```
def mostraImagemInvertida(caminho,nomeArquivo):
  import os
  import matplotlib.pyplot as plot
  from PIL import Image
  import numpy as np
  
  #concatena o caminho com nome no arquivo - revolve problemas de OS Win vs Linux
  arquivo = os.path.join(caminho,nomeArquivo)
  
  #cria uma instancia da image
  imgNormal = Image.open(arquivo)
  
  #inverte imagem - usando exemplo do livro "Automatize tarefas maçantes com Python, pg. 495"
  imgInvertida = imgNormal.transpose(Image.FLIP_LEFT_RIGHT)
  
  #converte a image para uma matriz
  img = np.array(imgInvertida)  
  
  #define qual escala a imagem deve ser exibida
  plot.figure(figsize=(10,5))
  
  
  
  #mostra a imagem
  plot.imshow(img)  

  
mostraImagemInvertida('.','fox.jpg')
```

#Criar uma função que retorne uma imagem invertida

```
def imagemInvertida(caminho,nomeArquivo):
  import os
  import matplotlib.pyplot as plot
  from PIL import Image
  import numpy as np
  
  #concatena o caminho com nome no arquivo - revolve problemas de OS Win vs Linux
  arquivo = os.path.join(caminho,nomeArquivo)
  
  #cria uma instancia da image
  imgNormal = Image.open(arquivo)
  
  #inverte imagem - usando exemplo do livro "Automatize tarefas maçantes com Python, pg. 495"
  imgInvertida = imgNormal.transpose(Image.FLIP_LEFT_RIGHT)
  
  return imgInvertida

```

#Criando um método para mostrar mais de uma imagem

```
def mostrarImagens(imagens,linhas=1,escala=(20,20)):
  import matplotlib.pyplot as plt
  if imagens is not None and isinstance(imagens,list) and len(imagens) >= 1:
    colunas = len(imagens)
  else:
    print('colecao de imagens invalida ou vazia')
    return
  fig, axes = plt.subplots(linhas, colunas, figsize=escala)
  axes = axes.flatten()
  for img, ax in zip( imagens, axes):
      ax.imshow(img)
  #plt.tight_layout()
  plt.show()
```
